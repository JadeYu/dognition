```python
%sql DESCRIBE dogs
```

```python
%%sql
SELECT dog_guid, gender, dog_fixed AS neutered, dimension AS personality,
    YEAR(updated_at)-birthday AS age, weight, breed, breed_type AS btype, 
    breed_group AS bgroup,total_tests_completed AS ntests, mean_iti_days AS mid
FROM dogs
WHERE exclude >0 AND gender IS NOT NULL AND dimension IS NOT NULL AND breed IS NOT NULL AND 
    total_tests_completed >0
```

```python
##inner join between complete test table and dogs table
```

```python
%%sql
SELECT c.dog_guid, c.created_at, DAYOFWEEK(c.created_at) AS day, MONTH(c.created_at) AS month, 
    d.gender, d.dog_fixed AS neutered, d.dimension AS personality,
    YEAR(d.updated_at)-d.birthday AS age, d.weight, d.breed, d.breed_type AS btype, 
    d.breed_group AS bgroup
FROM complete_tests c, dogs d
WHERE c.dog_guid = d.dog_guid AND d.exclude >0 AND d.gender IS NOT NULL AND d.dimension IS NOT NULL
    AND d.breed IS NOT NULL AND d.total_tests_completed >0
LIMIT 10
```

```python
#explort result
test_complete = %sql SELECT c.dog_guid, c.created_at, DAYOFWEEK(c.created_at) AS day, MONTH(c.created_at) AS month,  d.gender, d.dog_fixed AS neutered, d.dimension AS personality, YEAR(d.updated_at)-d.birthday AS age, d.weight, d.breed, d.breed_type AS btype, d.breed_group AS bgroup FROM complete_tests c, dogs d WHERE c.dog_guid = d.dog_guid AND d.exclude >0 AND d.gender IS NOT NULL AND d.dimension IS NOT NULL AND d.breed IS NOT NULL AND d.total_tests_completed >0
test_complete.csv("test_complete.csv")
```

```python
#Question 3: Does initiation (first test) date vary with dog features?
#Combined with data for Question 1, generate one table (one entry for each dog)
```

```python
%%sql
SELECT c.dog_guid, MIN(c.created_at), DAYOFWEEK(MIN(c.created_at)) AS day, MONTH(MIN(c.created_at)) AS month, 
    d.gender, d.dog_fixed AS neutered, d.dimension AS personality,
    YEAR(d.updated_at)-d.birthday AS age, d.weight, d.breed, d.breed_type AS btype, 
    d.breed_group AS bgroup, d.total_tests_completed AS ntests, d.mean_iti_days AS mid
FROM complete_tests c, dogs d
WHERE c.dog_guid = d.dog_guid AND d.exclude >0 AND d.gender IS NOT NULL AND d.dimension IS NOT NULL
    AND d.breed IS NOT NULL AND d.total_tests_completed >0
GROUP BY c.dog_guid
LIMIT 10
```

```python
dog_list = %sql SELECT c.dog_guid, MIN(c.created_at), DAYOFWEEK(MIN(c.created_at)) AS day, MONTH(MIN(c.created_at)) AS month,  d.gender, d.dog_fixed AS neutered, d.dimension AS personality, YEAR(d.updated_at)-d.birthday AS age, d.weight, d.breed, d.breed_type AS btype,  d.breed_group AS bgroup, d.total_tests_completed AS ntests, d.mean_iti_days AS mid FROM complete_tests c, dogs d WHERE c.dog_guid = d.dog_guid AND d.exclude >0 AND d.gender IS NOT NULL AND d.dimension IS NOT NULL AND d.breed IS NOT NULL AND d.total_tests_completed >0 GROUP BY c.dog_guid
dog_list.csv("dog_list.csv")
```
